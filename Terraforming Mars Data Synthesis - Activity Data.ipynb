```python
# Required imports

import pandas as pd
```

```python
df = pd.read_csv('/Users/akash/Desktop/Terraforming Mars Project/Atmosphere_Data.csv').iloc[:50000,1:]
```

```python
df = df.drop(['min_temp', 'max_temp', 'pressure', 'avg_temp', 'transformer', 'transformed_min_temp', 'transformed_max_temp', 'transformed_min_temp_dampened', 'transformed_max_temp_dampened', 'transformer_pressure'], axis = 1)
```

```python
df.columns = ['ls', 'sol', 'target_temp', 'target_pressure']
```

```python
import numpy as np
import random

df['activity'] = np.nan
df['activity'] = df.activity.apply(lambda x: random.choice(['Nuclear Bomb Seed', 'Radiation Cleaning', 'CO2 Pump', 'BioEngineered O2 Seed', 'Solar Focus', 'Launch Probe']) )

df['activity_type'] = np.nan
df['activity_type'] = df.activity.apply(lambda x: random.choice(['Level 1', 'Level 2', 'Level 3']) )

df['activity_handler'] = np.nan
df['activity_handler'] = df.activity.apply(lambda x: random.choice(['United Nations', 'ISRO', 'NASA', 'ESDA']) ) 
```

```python
df['earth_date'] = pd.date_range(start='1/1/2030', periods=len(df), freq='D')
```

```python
df['trigger_schedule_time'] = np.nan
df['trigger_schedule_time'] = df.trigger_schedule_time.apply(lambda x: random.choice(['12 AM', '12 PM']) ) 

df['mars_satellite_processing_delay_mins'] = np.nan
df['mars_satellite_processing_delay_mins'] = df.mars_satellite_processing_delay_mins.apply(lambda x: random.choice(range(0,10,1))) 
```

```python
locations = ['Columbia Hills, Gusev', 'Eberswalde',
'Holden Crater',
'Jezero Crater',
'Mawrth Vallis',
'NE Syrtis',
'Nili Fossae',
'SW Melas']
```

```python
df['martian_activity_location'] = np.nan
df['martian_activity_location'] = df.martian_activity_location.apply(lambda x: random.choice(locations)) 
```
